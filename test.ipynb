```python
import numpy as np
import matplotlib.pyplot as plt

# Replace 'your_file.npy' with the actual path to your .npy file
file_path = 'C:/Users/user/Downloads/images_medseg.npy'

try:
    # Load the .npy file
    data = np.load(file_path)

    # Assuming data is a 4-dimensional array with shape (num_images, height, width, channels)
    num_images, height, width, channels = data.shape

    # Display the first few images
    num_images_to_display = 3
    for i in range(min(num_images, num_images_to_display)):
        plt.subplot(1, num_images_to_display, i + 1)
        plt.imshow(data[i].squeeze(), cmap='gray')  # Assuming images are grayscale, adjust cmap if needed
        plt.title(f'Image {i + 1}')
        plt.axis('off')

    plt.show()

except FileNotFoundError:
    print(f"File not found: {file_path}")

except Exception as e:
    print(f"An error occurred: {e}")
```
